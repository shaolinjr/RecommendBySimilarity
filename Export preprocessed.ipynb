```python
import pandas as pd
import numpy as np
import collections
import sklearn.preprocessing as preprocessing
```

```python
df_categ = pd.read_json('dados.json',encoding="utf-8")
df_categ['user_id'] = range(0,len(df_categ))
# df_categ.set_index(df_categ['user_id'], inplace=True)
df_categ.tail()
```

```python
# Transformando coluna de segmentos
segmentos_dummy = pd.get_dummies(df_categ['segmento'])
df_categ.drop('segmento',axis=1,inplace=True)
df_categ = df_categ.join(segmentos_dummy)
```

```python
df_categ.tail()
```

```python
# Transformando coluna tipo_user
label_encoder = preprocessing.LabelEncoder()
label_encoder.fit(df_categ['tipo_user'])
label_encoder.classes_
df_categ['tipo_user'] = label_encoder.transform(df_categ['tipo_user'])
```

```python
df_categ.tail()
```

```python
df_categ.tail()
```

```python
# Retirando linhas com habilidades em branco

for line in df_categ.itertuples():
    if line.habilidades == ['']:
        print (line)
        df_categ = df_categ[df_categ.user_id != line.user_id]
        print(50*"*")
```

```python
df_categ
```

```python
# Arrumando os indexes do df após deletar linhas com habilidades em branco
# Fizemos isso por causa do .join, que estava dando problema de index e gerando NaN
df_categ = df_categ.reset_index(drop=True)

mlb = preprocessing.MultiLabelBinarizer()
habilidades_col = mlb.fit_transform(df_categ['habilidades'])
 
habilidades_df = pd.DataFrame(habilidades_col,columns=mlb.classes_)
df_categ = df_categ.join(habilidades_df)

df_categ.tail()
```

```python
# Deletamos a coluna categorica de habilidades
df_categ.drop('habilidades',axis=1,inplace=True)
```

```python
# Export preprocessed data
csv = df_categ.to_csv(path_or_buf='./data_json_preprocessed.csv',encoding='utf-8')
```
